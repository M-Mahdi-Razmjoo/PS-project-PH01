## Simulation 1

```python
import random

n = 30
A = 0
for i in range(1, 1001):
    random.seed(400101272+i)
    list = []
    for j in range(1,n+1):
        chosen_number = random.randint(1,365)
        list.append(chosen_number)
    if len(list) != len(set(list)):
        A = A+1

print("Number of times A occurred = ",A)
```

## Simulation 2

```python
import random
import numpy as np
import matplotlib.pyplot as plt


A = 0
As = []
for j in range(1,101):
    for i in range(1, 1001):
        random.seed(400101272+i)
        list = []
        for i in range(1,j+1):
            chosen_number = random.randint(1,365)
            list.append(chosen_number)
        if len(list) != len(set(list)):
            A = A+1
    As.append(A)
    A=0

n = np.arange(1,101,1)
plt.plot(n,As)
plt.xlabel('n')
plt.ylabel('Number of times A occurred')
plt.show()
```

## Simulation 4

```python
import math
import matplotlib.pyplot as plt
import numpy as np

def probabilityOfA(n):
    A = 1-(math.factorial(n)*math.comb(365,n))/(365**n)
    return A

n = np.arange(1,366,1)
As = []
check = False
for i in range(1,366):
    As.append(probabilityOfA(i))
    if probabilityOfA(i)>=0.5 and check==False:
        the_point_n=i
        the_point_A=probabilityOfA(i)
        check=True

plt.plot(n,As,label='Probability of A')
plt.plot([the_point_n, the_point_n],[0, the_point_A],'r--')
plt.plot([0, the_point_n],[the_point_A, the_point_A],'r--')
plt.legend()
plt.ylabel('Probability')
plt.xlabel('n')
plt.show()
```

## Simulation 5

```python
import math
import matplotlib.pyplot as plt
import numpy as np

def probabilityOfA(n):
    A = 1-(math.factorial(n)*math.comb(365,n))/(365**n)
    return A

def approximateProbabilityOfA(n):
    A = 1-math.exp(-((n*(n-1))/(730)))
    return A

n = np.arange(1,366,1)
As = []
AAs = []
check = False
for i in range(1,366):
    As.append(probabilityOfA(i))
    AAs.append(approximateProbabilityOfA(i))

plt.plot(n,As,'.',label='Probability Of A')
plt.plot(n,AAs,label='Approximate Probability Of A')
plt.legend()
plt.ylabel('Probability')
plt.xlabel('n')
plt.show()
```

## Simulation 7

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

random_data = expon.rvs(scale=0.3, size=100000)
rounded_data = [ round(elem, 4) for elem in random_data ]

N=[]
x=0
n=0
for i in range(len(rounded_data)):
    x+=rounded_data[i]
    if x<1:
        n+=1
    else:
        N.append(n)
        n=0
        x=1-x

plt.hist(N,density=False, bins=40,edgecolor='black')
plt.show()

```

## Simulation 9

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon


def numberOfCalls(T,S):
    random_data = expon.rvs(scale=S, size=100000)
    rounded_data = [ round(elem, 4) for elem in random_data ]

    N=[]
    x=0
    n=0
    for i in range(len(rounded_data)):
        x+=rounded_data[i]
        if x<T:
            n+=1
        else:
            N.append(n)
            n=0
            x=T-x
    return N

fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.hist(numberOfCalls(5,0.5),density=False, bins=40,edgecolor='black')
ax1.set_title("$T=5 , \lambda = 0.5$")
ax2.hist(numberOfCalls(10,0.5),density=False, bins=40,edgecolor='black')
ax2.set_title("$T=10 , \lambda = 0.5$")
fig.set_dpi(150)
plt.show()
```

## Simulation 10

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy

fig,axs = plt.subplots(2, 2)

x1 = np.arange(-10,10,0.01)
y1 = []
for i in x1:
    y1.append((1/(math.sqrt(2)*math.pi))*(math.exp(-(i**2)/2)))

x2 = np.arange(-10,10,0.01)
y2 = []
for i in x2:
    y2.append(scipy.stats.norm.cdf(i))

x3 = np.arange(0,20,0.01)
y3 = []
for i in x3:
    y3.append(math.exp(-i))

x4 = np.arange(0,20,0.01)
y4 = []
for i in x4:
    y4.append(1-math.exp(-i))


axs[0,0].plot(x1,y1)
axs[0,0].set_title('PDF of Gaussian distribution')
axs[0,1].plot(x2,y2)
axs[0,1].set_title('CDF of Gaussian distribution')
axs[1,0].plot(x3,y3)
axs[1,0].set_title('PDF of Exponential distribution')
axs[1,1].plot(x4,y4)
axs[1,1].set_title('CDF of Exponential distribution')
fig.set_dpi(200)
fig.tight_layout()
```

## Simulation 13

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_excel('data.xls')

values = df.iloc[:3288,5]
X=[]
for i in range(3288):
    if values[i]<2500:
        X.append(values[i])

Xbar = round(sum(X),2)/len(X)
muhat = Xbar
lambdahat = 1/Xbar

sigma=0
for i in range(len(X)):
    sigma+=(X[i]-Xbar)**2
sigmahat = sigma/len(X)

fig, (ax1, ax2) = plt.subplots(1, 2)

random_data1 = np.random.normal(loc=muhat, scale=sigmahat, size=1000)
ax1.hist(random_data1,density=False, bins=40,edgecolor='black')
ax1.set_title("$\mathcal{N} (\hat{\mu}_{MLE},\hat{\sigma}_{MLE}^2)$")

random_data2 = expon.rvs(scale=lambdahat, size=1000)
ax2.hist(random_data2,density=False, bins=40,edgecolor='black')
ax2.set_title("$\exp (\hat{\lambda}_{MLE})$")

fig.set_dpi(150)
plt.show()
```
